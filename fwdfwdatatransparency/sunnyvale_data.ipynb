# Citation Racial/Gender Breakdown

```python
file = "fwdfwdatatransparency/2019 Citation Data.csv"
citations19 = pd.read_csv(file)

#set first row's labels to be column names
citations19.columns = ['Report Number', 'Date', 'Violation Code', 'race', 'Sex']
citations19 = citations19.iloc[1:]

#fill in empty cells
citations19['race'] = citations19['race'].fillna("Not Recorded")
citations19['Sex'] = citations19['Sex'].fillna("Not Recorded")

citations19['Sex'].replace(['Male', 'Female'], ['M', 'F'], inplace=True)

citations19
```

```python
#group by race, count number per race category
race19 = citations19.groupby(['race']).count()

#reset index
race19.reset_index(inplace=True)

#leave only Race and Count as the 2 columns
race19 = race19[['race', 'Report Number']]
race19.columns = ['Race', 'Count']
race19.name = "Citations By Race in 2019"

race19 = pd.DataFrame(race19)
race19
```

```python
#group by gender, count number per race category
gender19 = citations19.groupby(['Sex']).count()

#reset index
gender19.reset_index(inplace=True)

#leave only Gender and Count as the 2 columns
gender19 = gender19[['Sex', 'Report Number']]
gender19.columns = ['Gender', 'Count']
gender19.name = "Citations By Gender in 2019"

gender19 = pd.DataFrame(gender19)
gender19
```

```python
file = "fwdfwdatatransparency/2020 Citation Data.csv"
citations20 = pd.read_csv(file)

#set first row's labels to be column names
citations20.columns = ['Report Number', 'Date', 'Violation Code', 'race', 'Sex']
citations20 = citations20.iloc[1:]

##fill in empty cells 
citations20['race'] = citations20['race'].fillna("Not Recorded")
citations20['Sex'] = citations20['Sex'].fillna("Not Recorded")

citations20['Sex'].replace(['Male', 'Female'], ['M', 'F'], inplace=True)

citations20
```

```python
#group by race, count number per race category
race20 = citations20.groupby(['race']).count()

#reset index
race20.reset_index(inplace=True)

#leave only Race and Count as the 2 columns
race20 = race20[['race', 'Report Number']]
race20.columns = ['Race', 'Count']
race20.name = "Citations By Race in 2020"

race20 = pd.DataFrame(race20)
race20
```

```python
#group by gender, count number per race category
gender20 = citations20.groupby(['Sex']).count()

#reset index
gender20.reset_index(inplace=True)

#leave only Gender and Count as the 2 columns
gender20 = gender20[['Sex', 'Report Number']]
gender20.columns = ['Gender', 'Count']
gender20.name = "Citations By Gender in 2020"

gender20 = pd.DataFrame(gender20)
gender20
```

```python
#combine 2019 and 2020 tables into one large table
all_citations = pd.concat([citations19, citations20])
```

```python
#group by race, count number per race category
all_race = all_citations.groupby(['race']).count()

#reset index
all_race.reset_index(inplace=True)

#leave only Race and Count as the 2 columns
all_race = all_race[['race', 'Report Number']]
all_race.columns = ['Race', 'Count']
all_race.name = "Citations By Race in 2019 and 2020 Combined"

all_race = pd.DataFrame(all_race)
all_race
```

```python
#group by gender, count number per race category
all_gender = all_citations.groupby(['Sex']).count()

#reset index
all_gender.reset_index(inplace=True)

#leave only Gender and Count as the 2 columns
all_gender = all_gender[['Sex', 'Report Number']]
all_gender.columns = ['Gender', 'Count']
all_gender.name = "Citations By Gender in 2020"

all_gender = pd.DataFrame(all_gender)
all_gender
```

# Approximation of Likelihood To Be Arrested/Cited Based on Race

Based on the Los Altos released data breakdown of the ratio of arrests/citations of arrests vs. nonresidents, this will use the same ratio to approximate the breakdown of residents vs. non residents in the Sunnyvale data. This is because otherwise, there is no way to approximate the likelihood of being arrested/cited on the basis of race. 

Additionally, it is a non-trivial issue to approximate the likelihood of being arrested/cited because there is no way of knowing the exact demographics of those who pass through Sunnyvale. Thus, to best approximate, I used an amalgamation of all of the neighboring cities to approximate the demographics of non residents.

A note: I'm not entirely sure why the sum of the number of residents doesn't equal the population given, but I think it has to do with how people can be counted under two or more races, thus double/triple counting certain people. However, there's no clear way to remove the double counting, so I am keeping all the information in the table for the sake of transparency.

```python
sunnyvale_pop = 152703
sunnyvale = pd.DataFrame(
    {'Race': ['White', 'Black', 'American Indian/Alaskan', 'Asian', 'Native Hawaiian', 'Two or More', 'Hispanic'],
     'Number of Residents': [sunnyvale_pop*0.421, sunnyvale_pop*0.017, sunnyvale_pop*0.004, sunnyvale_pop*0.459, sunnyvale_pop*0.002, sunnyvale_pop*0.044, sunnyvale_pop*0.173]})
sunnyvale
```

# Conclusions

The conclusion this seems to draw is that Black people are 4.2 times as likely to be cited as White people, and that Hispanic people are 2.7 times as likely to be cited as White people. Black people are 10.9 times as likely to be arrested as White people, while Hispanic people are 3.8 times as likely to be arrested as White people. 

That being said, it is highly, highly unlikely this is a completely accurate statistical representation. There are a multitude of factors that contribute to the likely inaccuracy. 

1) Being unable to determine residents vs. non residents. This makes it impossible to correctly draw conclusions about the exact rate of arrests/citations, as explained at the top of this section.
2) The large number of people who were categorized as "Other" or "Unknown" or not recorded at all. 
3) The categorization of people as "Two or More" races in the census while also likely double counting them in other races. 

There are likely many other factors I have forgotten. Please feel free to reach out to me with other thoughts on how to improve this. 

The purpose of this is not at all to draw any decisive conclusions. Instead, this is merely intended to show how much more data we need to collect before any definitive conclusions can be drawn, as well as call for a general greater accountability of all of our police departments.
