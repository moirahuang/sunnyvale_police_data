## Gender Analysis

```python

file = "fwdfwdatatransparency/2019 Citation Data.csv"
citations19 = pd.read_csv(file)

#set first row's labels to be column names
citations19.columns = ['Report Number', 'Date', 'Violation Code', 'race', 'Sex']
citations19 = citations19.iloc[1:]

#fill in empty cells
citations19['race'] = citations19['race'].fillna("Not Recorded")
citations19['Sex'] = citations19['Sex'].fillna("Not Recorded")

citations19['Sex'].replace(['Male', 'Female'], ['M', 'F'], inplace=True)

citations19
```

```python
#group by gender, count number per race category
gender19 = citations19.groupby(['Sex']).count()

#reset index
gender19.reset_index(inplace=True)

#leave only Gender and Count as the 2 columns
gender19 = gender19[['Sex', 'Report Number']]
gender19.columns = ['Gender', 'Count']
gender19.name = "Citations By Gender in 2019"

gender19 = pd.DataFrame(gender19)
gender19
```
